```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
# xtrain,ytrain,xtest,ytest = 'Xtrain','Xtest','Ytrain','Ytest'
```

```python
model = LinearRegression()

model.fit(xtrain,ytrain) #mx+c
```

```python
#predict() # y = mx+c (9423.81532303*X + 24380.20147947369)

# model.predict(input(X))

ytrain_predicted = model.predict(xtrain) # ytrain_predicted
```
