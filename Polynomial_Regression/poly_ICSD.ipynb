```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
df = pd.read_csv("Ice_cream selling data.csv")
```

```python
df.info()
```

```python
X = df[['Temperature (°C)']]
y = df['Ice Cream Sales (units)']

xtrain,xtest,ytrain,ytest = train_test_split(X,y,random_state=42,train_size=0.8)
```

```python
df.corr()
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
'''
1)r2_score()
r2_score(actual_output,predicted_output)
2)model.score()
model.score(actual_input,actual_output)
Both are same
'''
```

```python
train_score = model.score(xtrain,ytrain)
test_score = model.score(xtest,ytest)
```

```python
print(f'''
      Train score : {train_score}
      Test score : {test_score}
      ''')
```

```python
ytrain_pred = model.predict(xtrain)
ytest_pred = model.predict(xtest)
```

```python
ytrain
```

```python
import matplotlib.pyplot as plt
plt.scatter(xtrain,ytrain)
plt.scatter(xtrain,ytrain_pred)
```

## model summary
`model is underfit`

why model is underfit ? 

Model(linear model) is too simple not able to capture the complex patterns(non-linear data).

Solution ?

PolynomialFeatures + LinearRegression



```python
poly = PolynomialFeatures(degree=2)
'''
fit_transform()
fit - learns the pattern 
transform -transform values from learned patterns(fit)
PolynomialFeatures()
How fit_transform() works in PolynomialFeatures ?
fit - based on actual input(features) it adds new features based
on degree(parameter)
transform - transforms the values for created(added) columns.
syntax
poly.fit_transform(xtrain)
'''
```

```python
xtrain_poly = poly.fit_transform(xtrain)
model.fit(xtrain_poly,ytrain)
```

```python
xtrain_poly
```

```python
xtrain
```

```python
train_score = model.score(xtrain_poly,ytrain)
# test_score = model.score(xtest,ytest)
```

```python
train_score
```

```python
ypred_train = model.predict(xtrain_poly)

plt.scatter(xtrain,ytrain)
plt.scatter(xtrain,ypred_train)
```

```python
xtest_poly = poly.transform(xtest)
```

```python
ypred_test = model.predict(xtest_poly)

plt.scatter(xtest,ytest)
plt.scatter(xtest,ypred_test)
```
