```python
import pandas as pd
from sklearn.linear_model import Ridge,Lasso,ElasticNet,LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error
```

```python
''' 
Regularization types or parameters
What is weights ?
It tells the importance of features(columns)
Lasso ?(L1) |w|
it pushes the weights towards to zero,and exactly zero
if its weights are exactly zero those columns are not important
when to use Lasso ?
    -When we have more features and model is overfit
Ridge(L2) ? w²
It shrinks the weights but never force them to zero
when to use Ridge ?
When model is overfit and multicollinearity
ElasticNet(L1+L2) ?
It uses both lasso and ridge with the help of l1_ratio
it selects important features and also handles multicollinearity
when to use ElasticNet?
    when feature selection(more no of features) and 
    multicollinearity problem occurs.
    when ridge and lasso fails in overfit
cost function ?
MSE is a cost function in regression which calculates errors 
during model training
'''
```

```python
df = pd.read_csv('Student_Performance.csv')
```

```python
df.head()
```

```python
df.corr(numeric_only=True)['Performance Index']
```

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'Yes':1,'No':0})
```

```python
df.drop(columns='Performance Index')
```

```python
# df = df.drop(columns='Previous Scores')
```

```python
X = df.drop(columns='Performance Index')
# X = df[['Hours Studied','Previous Scores']]
y = df['Performance Index']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,random_state=42,train_size=0.8)
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
#train data
#actual output =>ytrain
#predicted output => model.predict(xtrain)
ypred_train = model.predict(xtrain)
r2_score(ytrain,ypred_train)
```

```python
#test data
#actual output =>ytest
#predicted output => model.predict(xtest)
ypred_test = model.predict(xtest)
r2_score(ytest,ypred_test)
```

```python
model1  = Ridge()
model1.fit(xtrain,ytrain)
model1.coef_
```

```python
model1.score(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

```python
model2 = ElasticNet(l1_ratio=0.3)
model2.fit(xtrain,ytrain)
```
